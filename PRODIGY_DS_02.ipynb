# Handling and Detecting Outliers

```python
# Box plot for 'Age'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])
plt.title('Box Plot of Age')
plt.show()

# Box plot for 'Fare'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Fare'])
plt.title('Box Plot of Fare')
plt.show()
```

```python
# Define a function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers in 'Age'
age_outliers = detect_outliers_iqr(df, 'Age')
print(f'Outliers in Age column:\n{age_outliers}')

# Detect outliers in 'Fare'
fare_outliers = detect_outliers_iqr(df, 'Fare')
print(f'Outliers in Fare column:\n{fare_outliers}')
```

```python
# Scatter plot for 'Age' vs 'Fare'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

```python
# Remove outliers from 'Age'
age_Q1 = df['Age'].quantile(0.25)
age_Q3 = df['Age'].quantile(0.75)
age_IQR = age_Q3 - age_Q1
age_lower_bound = age_Q1 - 1.5 * age_IQR
age_upper_bound = age_Q3 + 1.5 * age_IQR
df_no_age_outliers = df[(df['Age'] >= age_lower_bound) & (df['Age'] <= age_upper_bound)]

# Remove outliers from 'Fare'
fare_Q1 = df['Fare'].quantile(0.25)
fare_Q3 = df['Fare'].quantile(0.75)
fare_IQR = fare_Q3 - fare_Q1
fare_lower_bound = fare_Q1 - 1.5 * fare_IQR
fare_upper_bound = fare_Q3 + 1.5 * fare_IQR
df_no_fare_outliers = df_no_age_outliers[(df_no_age_outliers['Fare'] >= fare_lower_bound) & (df_no_age_outliers['Fare'] <= fare_upper_bound)]

print(f'Dataset after removing outliers:\n{df_no_fare_outliers.head()}')
```
